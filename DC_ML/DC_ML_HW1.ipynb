Используя цикл for:

  
Используя любой из видов цикла - создать список названий вакансий, больших по зп в Санкт-Петербурге чем, в Москве. Для оценки зарплаты брать нижнюю границу (from), очищенную от налога (-13%), и пересчитанную с учетом курса валют в рубли. 



```python
import collections as cl
```

```python
items_msc = []
items_spb = []

# получаем 2000 вакансий по Москве и СПБ
for i in range(20):
  url_msc = f"https://api.hh.ru/vacancies?area=1&per_page=100&page={i}"
  items_msc += requests.get(url_msc).json()['items']
  url_spb = f"https://api.hh.ru/vacancies?area=2&per_page=100&page={i}"
  items_spb += requests.get(url_spb).json()['items']
```

```python
# словарь: вакансия с одним и тем же наименованием по Москве - индекс вакансии в исходном списке (нужно для того, чтобы определить максимальную зарплату для вакансии в Москве)
job_freq_msc = cl.defaultdict(list)
for i in range(len(items_msc)):
    job_freq_msc[items_msc[i]['name']].append(i)

job_freq_msc
```

```python
# словарь: вакансия с одним и тем же наименованием по СПБ - индекс вакансии в исходном списке (нужно для того, чтобы определить максимальную зарплату для вакансии в СПБ)
job_freq_spb = cl.defaultdict(list)
for i in range(len(items_spb)):
    job_freq_spb[items_spb[i]['name']].append(i)

job_freq_spb
```

```python
# названия вакансий, по которым мы будем проводить итоговое сравнение зарплаты
s1 = set(job_freq_msc)
s2 = set(job_freq_spb)
general_job_names = s1 & s2
general_job_names
```

```python
# словарь: валюта - курс
currencies = {}
dictionaries = requests.get('https://api.hh.ru/dictionaries').json()
for currency in dictionaries['currency']:
    currencies[currency['code']] = currency['rate']

# возвращает нижнюю границу зарплаты с вычетом налогов и пересчитанную с учетом курса валют в рубли
def compute_final_salary(currencies, salary_info):
  salary = 0

  if salary_info is not None:
    gross = 0.87 if salary_info['gross'] else 1

    if salary_info['from']:
      salary = salary_info['from']*gross/currencies[salary_info['currency']]

  return salary
```

```python
# определяем максимальную зарплату для вакансии по одному городу
# результирующий словарь для вакансий pre_result: название вакансии - [макс. зарплата по Москве, макс. зарплата по СПБ]
pre_result = {}
for vac in general_job_names:

  # по Москве
  msc_sal = [compute_final_salary(currencies, items_msc[index]['salary']) for index in job_freq_msc[vac]]
  pre_result[vac] = [max(msc_sal)]

  # по СПБ
  spb_sal = [compute_final_salary(currencies, items_spb[index]['salary']) for index in job_freq_spb[vac]]
  pre_result[vac].append(max(spb_sal))

pre_result
```

```python
# создаем список названий вакансий, больших по зп в Санкт-Петербурге чем, в Москве
# если одно из чисел 0 (зарплата не указана), то вакансию пропускаем
result_spb = {}
for k, v in pre_result.items():
  if v[0] < v[1] and v[0] != 0 and v[1] != 0:
    result_spb[k] = v[1]

result_vac_list_spb = [key for key in result_spb.keys()]
result_vac_list_spb
```

```python
print(len(result_vac_list_spb))
print(max(result_spb, key=result_spb.get), max(result_spb.values()), sep=" -> ")
```
